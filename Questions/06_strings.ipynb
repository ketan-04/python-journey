#### ### **1. Find the length of a string**

#### Ask the user to enter a string and print how many characters it has using `len()`.

```python
st= input("enter string:")
print(len(st))
```

#### ### **2. Convert a string to uppercase**

#### Take a string input and print it in **uppercase** using `.upper()`.

```python
st1 = input("enter a string in lowercase :")
print(st1.upper())
```

#### ### **3. Count how many times a character appears in a string**

##### Ask the user to enter a string and a character. Count how many times the character appear

```python
st2 = input("enter string to check:")
char= input("enter character to search in string:")
check= st2.count(char)
print(check)
```
