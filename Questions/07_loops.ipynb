#### **1. Print numbers from 1 to 5 using `range()`**

#### Use a `for` loop and `range()` to print numbers from 1 to 5.

#### **Input:**

#### List of numbers: `range(1, 6)`


```python
for i in range(1,6):
    print(i)
```

### **2. Print even numbers from 1 to 10**

#### Loop through numbers from 1 to 10. Print only even numbers using a condition.

#### **Input:**

#### Numbers: `1 to 10`

```python
for i in range(1,11):
    if i % 2 == 0 :
        print(i)
```

### **4. Print each character of a string on a new line**

#### Use a loop to print each letter of the string.

#### **Input:**

#### String: `"hello"`

```python
st= "hello"
for i in st:
    print(i)
```

### **6. Count how many times a character appears in a string using loop**

#### Use a loop and condition to count a specific character.

#### **Input:**

#### String: `"banana"`

#### Character to count: `'a'`

```python
st1="banana"
co=0
for i in st1:
    if i == 'a':
        co+=1
print("thw word'a' appears ",co,"times")
```
