### **1. Check if a number is even or odd**

### Ask the user to enter a number. Print whether it is **even** or **odd**.

```python
num =  int(input("enter number to check:"))
if num %2==0:
    print("EVEN")
else : 
    print("ODD")
```

### **2. Check if a number is positive, negative, or zero**

### Take a number from the user and classify it as **positive**, **negative**, or **zero**.

```python
num1= int(input("enter number to be check:"))
if num1<0:
    print("Negative")
elif num1==0:
    print("Zero")
else:
    print("Positive")
```

#### **3. Voting eligibility based on age**

#### Ask the user to enter their **age**. If the age is **18 or more**, print “Eligible to vote”; otherwise, print “Not eligible to vote”.

```python
age = int(input("enter your age:"))
if age>18:
    print("Eligible to vote")
else:
    print("Not Eligible to vote")
```

#### 5.**Vowel or consonant check**

#### Ask the user to enter a **single character**. Check if it’s a **vowel** or **consonant**.

```python
char = input("enter single character")
s_vowels = "aeiou"

if char in s_vowels :
    print("VOWELS")
else:
    print("CONSONANTS")
```

#### **Grade based on marks**

#### Take **marks out of 100** and print the **grade**:

- **90+** → Grade A
- **75–89** → Grade B
- **60–74** → Grade C
- **Below 60** → Grade D

```python
marks = int(input("enter your marks:"))
if marks>90 :
    print("A")
elif marks <=90 and marks >= 75:
    print("B")
elif marks>= 60  and marks<= 74:
    print("C")
else:
    print("D")
```
