Given two sets A and B, print: Union, Intersection, A-B, B-A, Symmetric Difference.

A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

```python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print(a.union(b))
print(a.intersection(b))
print(a.difference(b))
print(b.difference(a))
print(a.symmetric_difference(b))
```

Count how many unique words exist in a given sentence.

Input sentence:
"apple banana apple mango banana orange"

```python
s= "apple banana apple mango banana orange"
a=s.split()
# print(set(a))
b=set(a)
print(b)
print(len(b))
```

Given a list of numbers, find which numbers appear in BOTH list1 and list2.

Use intersection.

list1 = [1,2,3,4,5]
list2 = [4,5,6,7]

```python
list1 = [1,2,3,4,5]
list2 = [4,5,6,7]
a=set(list1)
b=set(list2)
print(a.intersection(b))
```
